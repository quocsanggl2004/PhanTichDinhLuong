```python
import json
import pandas as pd
import re
```

```python
with open('TLP.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
camera_video_df = df[['product_id', 'camera_video']].copy()
camera_video_df.to_csv('camera_video.csv', index=False, encoding='utf-8-sig')
```

```python
df = df.replace('Đang cập nhật', 'N/A')
```

```python
# name
df['name'] = df['name'].apply(lambda x: re.split(r'[|,|-]', x)[0].strip())
```

```python
# price
df['price'] = df['price'].apply(lambda x: 'isnull' if x == 0 else x)
min_price = df['price'][df['price'] != 'isnull'].min()
df['price'] = df['price'].apply(lambda x: min_price if x == 'isnull' else x)
```

```python
# special_price
df['special_price'] = df['special_price'].apply(lambda x: x if x != 0 else 'isnull')
df['special_price'] = df.apply(lambda row: row['price'] if row['special_price'] == 'isnull' else row['special_price'], axis=1)
```

```python
print(df['price'].dtype)
print(df['special_price'].dtype)
```

```python
# operating_system
df['operating_system'] = df['operating_system'].apply(lambda x: 'isnull' if x == 'N/A' else x)
most_common_os = df['operating_system'][df['operating_system'] != 'isnull'].mode()[0]
df['operating_system'] = df['operating_system'].apply(lambda x: most_common_os if x == 'isnull' else x)
```

```python
# os_version
df['os_version'] = df['os_version'].apply(lambda x: re.split(r',|hoặc cao hơn', x)[0].strip() if isinstance(x, str) else x)
df['os_version'] = df['os_version'].apply(lambda x: 'isnull' if x == 'N/A' else x)
most_common_os_version = df['os_version'][df['os_version'] != 'isnull'].mode()[0]
df['os_version'] = df['os_version'].apply(lambda x: most_common_os_version if x == 'isnull' else x)
```

```python
df
```

```python
# Lọc dữ liệu cột battery, chỉ lấy đoạn dữ liệu có chứa 'mAh' sau đó mới phân tích, loại bỏ đơn vị 
df['battery'] = df['battery'].apply(lambda x: x if isinstance(x, str) and 'mAh' in x else '0 mAh')
df['battery'] = df['battery'].apply(lambda x: x.split('mAh')[0] if isinstance(x, str) else x)

def clean_battery(battery):
    if pd.isna(battery):
        return 0
    # Loại bỏ đơn vị mAh
    battery = re.sub(r'[^\d,]', '', str(battery))
    # Chuyển đổi giá trị từ dạng 5,232 thành 5232
    battery = battery.replace(',', '')
    battery = battery.replace('.', '')
    return int(battery) if battery else 0

df['battery'] = df['battery'].apply(clean_battery)

# Thay thế giá trị 0 bằng 'isnull'
df['battery'] = df['battery'].apply(lambda x: 'isnull' if x == 0 else x)

# Tìm giá trị min của cột battery (không bao gồm 'isnull') và làm tròn về int
min_battery = int(df['battery'][df['battery'] != 'isnull'].min())

# Thay thế 'isnull' bằng giá trị min đã làm tròn
df['battery'] = df['battery'].apply(lambda x: min_battery if x == 'isnull' else x)

df
```

```python
print(df['battery'].dtype)
```

```python
# chipset
df['chipset'] = df['chipset'].apply(lambda x: x.split(',')[0] if isinstance(x, str) else x)
df['chipset'] = df['chipset'].apply(lambda x: re.sub(r'\s*\d+nm|\s*tiến trình nm|\s*\(.*?\)', '', str(x), flags=re.IGNORECASE))
df['chipset'] = df['chipset'].replace('Đang cập nhật', 0)
df['chipset'] = df['chipset'].apply(lambda x: x.replace('thế hệ', 'Gen') if isinstance(x, str) else x)
df['chipset'] = df['chipset'].apply(lambda x: re.sub(r'[^\w\s]', '', x) if isinstance(x, str) else x)
df['chipset'] = df['chipset'].astype(object)
df
```

```python
# Thay thế giá trị N/A thành không ở cột chipset
df['chipset'] = df['chipset'].replace(['NA', ''], 'Không')
df
```

```python
# display_size
df['display_size'] = df['display_size'].apply(lambda x: re.sub(r'[^\d.]', '', str(x)) if isinstance(x, str) else '0').replace('', '0').astype(float)
min_display_size = df['display_size'][df['display_size'] != 0].min()
df['display_size'] = df['display_size'].apply(lambda x: min_display_size if x == 0 else x)
df
```

```python
print(df['display_size'].dtype)
```

```python
# mobile_type_of_display
df['mobile_type_of_display'] = df['mobile_type_of_display'].replace('N/A', 'isnull')
df['mobile_type_of_display'] = df['mobile_type_of_display'].replace('isnull', 'Không')
```

```python
# mobile_cam_bien_van_tay
df['mobile_cam_bien_van_tay'] = df['mobile_cam_bien_van_tay'].replace('N/A', 'isnull')
df['mobile_cam_bien_van_tay'] = df['mobile_cam_bien_van_tay'].replace('isnull', 'Không')
```

```python
# warranty_information
df['warranty_information'] = df['warranty_information'].replace('N/A', 'isnull')
df['warranty_information'] = df['warranty_information'].replace('isnull', 'Không')
```

```python
# key_selling_points
df['warranty_information'] = df['warranty_information'].replace('N/A', 'isnull')
df['warranty_information'] = df['warranty_information'].replace('isnull', 'Không')
```

```python
# promotion_information
df['promotion_information'] = df['promotion_information'].replace('N/A', 'isnull')
df['promotion_information'] = df['promotion_information'].replace('isnull', 'Không')
```

```python
# dislkplay_resolution
df['display_resolution'] = df['display_resolution'].apply(lambda x: re.split(r' pixels', x)[0] if isinstance(x, str) else x)
```

```python
def calculate_product(resolution):
    try:
        parts = resolution.split(' x ')
        if len(parts) == 2:
            return int(parts[0]) * int(parts[1])
    except:
        return 'isnull'
    return 'isnull'

# Áp dụng hàm cho cột display_resolution
df['display_resolution'] = df['display_resolution'].apply(calculate_product)

# Tìm giá trị nhỏ nhất không bao gồm 'isnull'
min_display_resolution = df['display_resolution'][df['display_resolution'] != 'isnull'].min()

# Thay thế 'isnull' bằng giá trị nhỏ nhất
df['display_resolution'] = df['display_resolution'].apply(lambda x: min_display_resolution if x == 'isnull' else x)

df
```

```python
# memory_internal
df['memory_internal'] = df['memory_internal'].apply(lambda x: int(re.sub(r'[^\d]', '', x.split('+')[0])) if isinstance(x, str) and re.sub(r'[^\d]', '', x.split('+')[0]) != '' else 0)
df['memory_internal'] = df['memory_internal'].astype(int)

# Thay thế giá trị 0 bằng 'isnull'
df['memory_internal'] = df['memory_internal'].apply(lambda x: 'isnull' if x == 0 else x)

# Tìm giá trị trung vị của cột memory_internal (không bao gồm 'isnull')
median_memory_internal = df['memory_internal'][df['memory_internal'] != 'isnull'].median()

# Thay thế 'isnull' bằng giá trị trung vị
df['memory_internal'] = df['memory_internal'].apply(lambda x: median_memory_internal if x == 'isnull' else x)
```

```python
# storage
# Kiểm tra và chuyển đổi đơn vị từ TB sang GB trước khi lọc và phân tích
def convert_tb_to_gb(storage):
    if isinstance(storage, str) and 'TB' in storage:
        # Lấy giá trị số và chuyển đổi từ TB sang GB
        return int(re.sub(r'[^\d]', '', storage)) * 1024
    return storage

df['storage'] = df['storage'].apply(convert_tb_to_gb)

# Lọc dữ liệu cột storage, chỉ lấy đoạn dữ liệu có chứa số, sau đó mới phân tích
df['storage'] = df['storage'].apply(lambda x: int(re.sub(r'[^\d]', '', str(x))) if isinstance(x, str) and re.sub(r'[^\d]', '', str(x)) else x)
```

```python
# storage
df['storage'] = df['storage'].replace('N/A', 'isnull')

# Tìm giá trị trung vị của cột storage (không bao gồm 'isnull')
median_storage = df['storage'][df['storage'] != 'isnull'].median()

# Thay thế 'isnull' bằng giá trị trung vị
df['storage'] = df['storage'].apply(lambda x: median_storage if x == 'isnull' else x).astype(int)
```

```python
# product_weight
df['product_weight'] = df['product_weight'].apply(lambda x: re.sub(r'[^\d]', '', x.split('g')[0]) if isinstance(x, str) else x)
df['product_weight'] = df['product_weight'].apply(lambda x: int(x) if x else 0)
# Chuyển đổi các giá trị lớn hơn 1000 về giá trị thập phân và ép kiểu float
df['product_weight'] = df['product_weight'].apply(lambda x: x / 10 if x > 1000 else x).astype(float)
df
```
