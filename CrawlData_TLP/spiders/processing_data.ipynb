```python
import json
import pandas as pd
import re
```

```python
with open('TLP.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Thay thế dữ liệu trống và dữ liệu N/A bằng giá trị 0
df = df.replace('N/A', 0)
df = df.fillna(0)
df
```

```python
print(df['mobile_ra_mat'].dtype)
```

```python
df['price'] = df['price'].apply(lambda x: 'N/A' if x == 0 else x)
df['special_price'] = df['special_price'].apply(lambda x: 'N/A' if x == 0 else x)
```

```python
print(df['price'].dtype)
print(df['special_price'].dtype)
```

```python
# battery
df['battery'] = df['battery'].apply(lambda x: x if 'mAh' in str(x) else 0)
# Lọc dữ liệu cột battery, chỉ lấy đoạn dữ liệu trước chữ 'mAh' sau đó mới phân tích
df['battery'] = df['battery'].apply(lambda x: x.split('mAh')[0] if isinstance(x, str) else x)
def clean_battery(battery):
    if pd.isna(battery):
        return 0
    # Loại bỏ đơn vị mAh
    battery = re.sub(r'[^\d,]', '', str(battery))
    # Chuyển đổi giá trị từ dạng 5,232 thành 5232
    battery = battery.replace(',', '')

    battery = battery.replace('.', '')
    return int(battery) if battery else 0

df['battery'] = df['battery'].apply(clean_battery)
df['battery'] = df['battery'].astype(int)
df
```

```python
print(df['battery'].dtype)
```

```python
# chipset
df['chipset'] = df['chipset'].apply(lambda x: x.split(',')[0] if isinstance(x, str) else x)
df['chipset'] = df['chipset'].apply(lambda x: re.sub(r'\s*\d+nm|\s*tiến trình nm|\s*\(.*?\)', '', str(x), flags=re.IGNORECASE))
df['chipset'] = df['chipset'].replace('Đang cập nhật', 0)
df['chipset'] = df['chipset'].apply(lambda x: x.replace('thế hệ', 'Gen') if isinstance(x, str) else x)
df['chipset'] = df['chipset'].apply(lambda x: re.sub(r'[^\w\s]', '', x) if isinstance(x, str) else x)
df['chipset'] = df['chipset'].astype(object)
df
```

```python
# display_size
df['display_size'] = df['display_size'].apply(lambda x: re.sub(r'[^\d.]', '', x) if x else '0').replace('', '0').astype(float)
df

df['display_size'] = df['display_size'].astype(float)
```

```python
print(df['display_size'].dtype)
```

```python
# display_resolution
df['display_resolution'] = df['display_resolution'].apply(lambda x: re.sub(r'\s*\(.*?\)|[^\d×x]', '', x) if isinstance(x, str) else x)
df['display_resolution'] = df['display_resolution'].apply(lambda x: x[:-1] if isinstance(x, str) and x.endswith('x') else x)
df
```

```python
# memory_internal
df['memory_internal'] = df['memory_internal'].apply(lambda x: int(re.sub(r'[^\d]', '', x.split('+')[0])) if isinstance(x, str) and re.sub(r'[^\d]', '', x.split('+')[0]) else 0)
df['memory_internal'] = df['memory_internal'].astype(int)
```

```python
print(df['memory_internal'].dtype)
```

```python
# storage
df['storage'] = df['storage'].apply(lambda x: int(re.sub(r'[^\d]', '', x)) if isinstance(x, str) and re.sub(r'[^\d]', '', x) else 0)
df['storage'] = df['storage'].astype(int)
```

```python
print(df['storage'].dtype)
```

```python
# product_weight
df['product_weight'] = df['product_weight'].apply(lambda x: re.sub(r'[^\d]', '', x.split('g')[0]) if isinstance(x, str) else x)
df['product_weight'] = df['product_weight'].apply(lambda x: int(x) if x else 0)
# Chuyển đổi các giá trị lớn hơn 1000 về giá trị thập phân và ép kiểu float
df['product_weight'] = df['product_weight'].apply(lambda x: x / 10 if x > 1000 else x).astype(float)
df
```

```python
print(df['product_weight'].dtype)
```

```python
# mobile_tan_so_quet
df['mobile_tan_so_quet'] = df['mobile_tan_so_quet'].apply(lambda x: int(re.sub(r'[^\d]', '', x)) if isinstance(x, str) and re.sub(r'[^\d]', '', x) else 0)
```

```python
print(df['mobile_tan_so_quet'].dtype)
```

```python
# total_count
df['total_count'] = df['total_count'].astype(int)
```

```python
print(df['total_count'].dtype)
```

```python
# average_rating
df['average_rating'] = df['average_rating'].astype(float)
```

```python
print(df['average_rating'].dtype)
```

```python
# Đổi tên các cột
df.rename(columns={
    'product_id': 'product_id',
    'name': 'name',
    'mobile_ra_mat': 'mobile_ra_mat',
    'price': 'price (VND)',
    'special_price': 'special_price (VND)',
    'manufacturer': 'manufacturer',
    'operating_system': 'operating_system',
    'os_version': 'os_version',
    'battery': 'battery (mAh)',
    'chipset': 'chipset',
    'display_size': 'display_size (inch)',
    'display_resolution': 'display_resolution (px)',
    'mobile_type_of_display': 'mobile_type_of_display',
    'camera_primary': 'camera_primary',
    'camera_secondary': 'camera_secondary',
    'camera_video': 'camera_video',
    'memory_internal': 'memory_internal (GB)',
    'storage': 'storage (GB)',
    'product_weight': 'product_weight (g)',
}, inplace=True)

df
```

```python
# Chuyển toàn bộ dữ liệu 0 thành N/A
df['chipset'] = df['chipset'].replace([0, '0'], 'N/A')
df = df.replace(0, 'N/A')
df = df.replace('Đang cập nhật', 'N/A')
df
```

```python
df.to_csv('processed_data.csv', index=False, encoding='utf-8-sig')
```
