```python
from psycopg2.extras import RealDictCursor
import pyodbc
import psycopg2

# Kết nối tới PostgreSQL với RealDictCursor
postgres_connection = psycopg2.connect(
    dbname='mydatabase',
    user='postgres',
    password='141216',
    host='localhost',
    port='5432'
)
postgres_cursor = postgres_connection.cursor(cursor_factory=RealDictCursor)

# Kết nối tới SQL Server
sqlserver_connection = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=KID1412\SQLEXPRESS;'
    'DATABASE=PTDL_CK;'
    'Trusted_Connection=yes;'
)
sqlserver_cursor = sqlserver_connection.cursor()

try:
    # Lấy dữ liệu từ PostgreSQL
    postgres_cursor.execute("SELECT * FROM telephone")
    rows = postgres_cursor.fetchall()

    # Kiểm tra dữ liệu
    if not rows:
        print("Bảng telephone trong PostgreSQL trống, không có dữ liệu để chuyển.")
    else:
        print(f"Đã lấy {len(rows)} bản ghi từ PostgreSQL.")

    # Chèn dữ liệu vào SQL Server
    insert_query = """
    INSERT INTO telephone (
        product_id, name, price, special_price, total_count, manufacturer, 
        operating_system, os_version, battery, chipset, display_size, 
        display_resolution, mobile_type_of_display, memory_internal, 
        storage, product_weight, mobile_tan_so_quet, loai_mang, mobile_cong_sac, 
        mobile_cam_bien_van_tay, warranty_information, average_rating, 
        key_selling_points, promotion_information
    ) 
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """
    
    for row in rows:
        print(f"Đang xử lý row: {row}")  # In từng row để kiểm tra kiểu dữ liệu
        
        # Kiểm tra xem product_id đã tồn tại trong bảng telephone chưa
        check_query = "SELECT COUNT(*) FROM telephone WHERE product_id = ?"
        sqlserver_cursor.execute(check_query, (row['product_id'],))
        count = sqlserver_cursor.fetchone()[0]
        
        if count == 0:
            # Chèn dữ liệu vào nếu chưa tồn tại
            sqlserver_cursor.execute(insert_query, (
                row['product_id'], row['name'], row['price'], row['special_price'], row['total_count'],
                row['manufacturer'], row['operating_system'], row['os_version'], row['battery'], row['chipset'],
                row['display_size'], row['display_resolution'], row['mobile_type_of_display'], row['memory_internal'],
                row['storage'], row['product_weight'], row['mobile_tan_so_quet'], row['loai_mang'], row['mobile_cong_sac'],
                row['mobile_cam_bien_van_tay'], row['warranty_information'], row['average_rating'],
                row['key_selling_points'], row['promotion_information']
            ))
            print(f"Chèn product_id {row['product_id']} vào SQL Server.")
        else:
            print(f"product_id {row['product_id']} đã tồn tại, bỏ qua.")

    # Thực hiện commit sau khi chèn tất cả dữ liệu
    sqlserver_connection.commit()

    print("Dữ liệu đã được chuyển từ PostgreSQL sang SQL Server thành công!")

except Exception as e:
    print(f"Lỗi xảy ra: {e}")
    sqlserver_connection.rollback()  # Rollback nếu có lỗi xảy ra

# Đóng kết nối
postgres_cursor.close()
postgres_connection.close()
sqlserver_cursor.close()
sqlserver_connection.close()
```

```python
import pandas as pd
import psycopg2
import pyodbc

# Kết nối đến PostgreSQL
postgres_connection = psycopg2.connect(
    dbname='mydatabase',   # Tên database PostgreSQL của bạn
    user='postgres',       # Tên người dùng PostgreSQL
    password='141216',     # Mật khẩu người dùng
    host='localhost',      # Địa chỉ host
    port='5432'            # Cổng kết nối (mặc định là 5432)
)

postgres_cursor = postgres_connection.cursor()

# Kết nối đến SQL Server
sql_server_connection = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=KID1412\SQLEXPRESS;'
    'DATABASE=PTDL_CK;'
    'Trusted_Connection=yes;'                        # Mật khẩu người dùng
)

sql_server_cursor = sql_server_connection.cursor()

# Nhập dữ liệu từ bảng Products trong PostgreSQL
try:
    # Lấy dữ liệu từ PostgreSQL
    postgres_cursor.execute("SELECT * FROM Products;")
    rows = postgres_cursor.fetchall()

    # Tạo DataFrame từ dữ liệu nhận được
    columns = [desc[0] for desc in postgres_cursor.description]  # Lấy tên cột từ description
    df = pd.DataFrame(rows, columns=columns)

    # Xóa dữ liệu cũ trong SQL Server (nếu cần)
    sql_server_cursor.execute("DELETE FROM Products;")  # Giả sử bảng trong SQL Server cũng có cùng tên
    sql_server_connection.commit()  # Lưu thay đổi

    # Chèn dữ liệu vào SQL Server
    for index, row in df.iterrows():
        insert_query = """
        INSERT INTO Products (product_id, name, price, special_price, total_count, manufacturer_id, os_id, display_id, battery, chipset, memory_internal, storage, product_weight, mobile_tan_so_quet, loai_mang, mobile_cong_sac, mobile_cam_bien_van_tay, warranty_information, average_rating, key_selling_points, promotion_information) 
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?) 
        """
        sql_server_cursor.execute(insert_query, (
            row['product_id'], row['name'], row['price'], row['special_price'], 
            row['total_count'], row['manufacturer_id'], row['os_id'], row['display_id'], 
            row['battery'], row['chipset'], row['memory_internal'], row['storage'], 
            row['product_weight'], row['mobile_tan_so_quet'], row['loai_mang'], 
            row['mobile_cong_sac'], row['mobile_cam_bien_van_tay'], 
            row['warranty_information'], row['average_rating'], 
            row['key_selling_points'], row['promotion_information']
        ))

    sql_server_connection.commit()  # Lưu thay đổi sau khi chèn dữ liệu thành công

    print("Data transferred successfully from PostgreSQL to SQL Server.")

except Exception as e:
    print(f"Error transferring data: {e}")

finally:
    # Đóng kết nối
    postgres_cursor.close()
    postgres_connection.close()
    sql_server_cursor.close()
    sql_server_connection.close()
```
